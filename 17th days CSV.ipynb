```python
s =" "
b =open('data.csv','w')
b.write("Name,Phone,age,add\n")
n =int(input("Enter how many rows ="))
for i in range(n):
    name =input("\nEnter the name=")
    add =input("Enter the address=")
    age =int(input("Enter age:"))
    phone =int(input("Enter the number="))
    info =f"{name},{phone},{age},{add}\n"
    s =s+info

b.write(s)
b.close()
```
